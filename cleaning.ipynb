```python
df.to_csv("GenZ_DatingApp_Cleaned.csv", index=False)
```

```python
# Function to convert "Daily Usage Time" to numeric (in hours)
def convert_usage_time(time_str):
    if pd.isna(time_str):
        return None  # Handle missing values
    elif "hour" in time_str:
        return float(time_str.split()[0])  # Extract the number before "hour"
    elif "minutes" in time_str or "minute" in time_str:
        return float(time_str.split()[0]) / 60  # Convert minutes to hours
    else:
        return None  # Handle unexpected cases
```

```python
df["Daily_Usage_Hours"] = df["Daily_Usage_Time"].apply(convert_usage_time)
```

```python
df.drop(columns=["Daily_Usage_Time"], inplace=True)
```
